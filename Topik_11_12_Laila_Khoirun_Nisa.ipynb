```python
#for uploading file
from google.colab import files

#for reading file
import io

#for importing file
import pandas as pd

#for visualization
import seaborn as sns

uploaded= files.upload()
df=pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
```

Dataset di atas terdiri dari 7043 data dengan jumlah kolom sebanyak 21. Untuk tipe data dari setiap variabel dapat dilihat pada **Dtype**, tipe data float64 berjumlah 1, int64 sebanyak 2, dan object berjumlah 18.

```python
# Mencari unique value variabel Churn
pd.unique(df['Churn'])
```

```python
# Mencari unique value variabel Senior Citizen
pd.unique(df['SeniorCitizen'])
```

Nilai 0 artinya bukan Senior Citizen, sedangkan nilai 1 artinya Senior Citizen

```python
# Jumlah Senior Citizen berdasarkan Churn di dataset
sns.countplot(data=df,x='SeniorCitizen',hue='Churn')
```

```python
# Cleaning inconsistency in data
# Mengubah data type Total Charges dari object menjadi float
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors="coerce")
df.info()
```

```python
# Check Duplicate
df[df.duplicated()]
```

Dataset yang digunakan tidak terdapat data duplikasi

```python
# Drop Duplicate
df=df.drop_duplicates()
```

```python
# Check data null
df[df.isna().any(axis=1)]
```

Terdapat 11 baris data null di kolom Total Charges

```python
# Check data null (dengan cara lain)
df.isnull().sum()
```

```python
# Drop data null
df=df.dropna()
```

```python
# Cek kembali data null setelah dilakukan drop data
df[df.isna().any(axis=1)]
```

Dataset tidak ada yang data null

```python
# Drop irrelevant data from analysis
df=df.drop(columns=['customerID'])
```

```python
# Get dummies to transform data 

# Can only be done on numerical data
# We need to convert all categorical data into 0 and 1 value

df2=pd.get_dummies(df)
df2
```

```python
df2.info()
```

```python
# Check Correlation

# 0 no correlation
# 1 positive correlation
# -1 negative correlation

df3=df2.corr()['Churn_Yes'].sort_values().reset_index()
df3
```

```python
# Korelasi antar variabel
sns.heatmap(df2.corr(), cmap="Blues")
sns.set(rc={'figure.figsize':(14,10)})
```

```python
# Drop Churn_Yes
df3=df3[~df3['index'].str.contains('Churn_Yes')]
```

```python
chart = sns.barplot(data=df3,x="index",y="Churn_Yes")

for item in chart.get_xticklabels():
  item.set_rotation(90)
```
