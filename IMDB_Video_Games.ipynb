```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/sample_data/imdb-videogames.csv")
data
```

```python
mean = data['rating'].mean()
median = data['rating'].median()
mode = data['rating'].mode()
print(f"Mean = {mean}\nMedian = {median}\nModus = {mode}")
```

```python
rating1 = data['rating']
votes1 = data['votes']
print(list(zip(rating1,votes1)))
print(f"Correlation = {np.corrcoef(rating1,votes1)[0,1]}")
```

```python
data.boxplot(column='rating',by='certificate')
plt.xticks(rotation='vertical')
plt.show()
```

```python
data['year'].hist(bins=15)
plt.suptitle('Year distribution of IMDB Video Games')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```
